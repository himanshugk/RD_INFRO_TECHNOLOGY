```python
# ---------------------------------------------------
# Task 1: Data Collection (Real Retail Dataset)
# ---------------------------------------------------

import pandas as pd
import requests

# 1Ô∏è‚É£ Load the dataset
df = pd.read_csv("retail_sales_dataset.csv")
print("‚úÖ Retail data loaded successfully!")
print(df.head())

# 2Ô∏è‚É£ Fetch exchange rates from public API
url = "https://api.exchangerate-api.com/v4/latest/USD"
response = requests.get(url)
data = response.json()

# 3Ô∏è‚É£ Extract INR and JPY rates
usd_to_inr = data["rates"]["INR"]
usd_to_jpy = data["rates"]["JPY"]
print(f"‚úÖ USD‚ÜíINR: {usd_to_inr}, USD‚ÜíJPY: {usd_to_jpy}")

# 4Ô∏è‚É£ Add rate columns
df["USD_TO_INR"] = usd_to_inr
df["USD_TO_JPY"] = usd_to_jpy

# 5Ô∏è‚É£ Save merged dataset
df.to_csv("raw_sales_data.csv", index=False)
print("üìÅ raw_sales_data.csv created successfully!")
```
